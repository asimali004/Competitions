```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/train.csv")
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.head(2)
```

```python
df1 = df.iloc[:,[1,5,7,8,10,11]]
df1
```

```python
for a in df1.columns:
  print(a,len(df1[a].unique()))
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
df1 = ohe.fit_transform(df1)
```

```python
df1=pd.DataFrame(df1.toarray())
```

```python
df2 = df.iloc[:,[2,3,4,6,9,12,13]]
df2.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
df2 = sc.fit_transform(df2)
```

```python
df2 = pd.DataFrame(df2)
```

```python
final = pd.concat([df2,df1],axis=1)
```

```python
final
```

```python
y = df.iloc[:,14]
```

```python
x = final
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(x_train,y_train)
```

```python
y_pred2 = dtr.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(x_train,y_train)
```

```python
y_pred3 = rfr.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred3)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred4 = svr.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred4)
```
